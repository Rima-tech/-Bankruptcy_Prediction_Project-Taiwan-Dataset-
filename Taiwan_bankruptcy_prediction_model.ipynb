```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# 1. Train the Random Forest model
rf_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    class_weight='balanced'  # helps with class imbalance
)
rf_model.fit(X_train, y_train)

# 2. Predict on test set
y_pred_rf = rf_model.predict(X_test)

# 3. Evaluate performance
accuracy_rf = accuracy_score(y_test, y_pred_rf)
cm_rf = confusion_matrix(y_test, y_pred_rf)

print(f" Random Forest Accuracy: {accuracy_rf:.4f}")
print("Confusion Matrix:")
print(cm_rf)


# 5. Plot feature importances
importances = rf_model.feature_importances_
feature_names = X_train.columns

# Convert to DataFrame for easy sorting and visualization
feat_imp_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

# Plot Top 15 features
plt.figure(figsize=(10, 6))
sns.barplot(data=feat_imp_df.head(15), x='Importance', y='Feature')
plt.title("Top 15 Important Features - Random Forest")
plt.tight_layout()
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay, classification_report
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt

# Step 1: Apply SMOTE to balance the training set
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Step 2: Train the kNN model on the resampled data
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_resampled, y_train_resampled)

# Step 3: Predict on original test set
y_pred_knn = knn.predict(X_test)

# Step 4: Evaluate
accuracy_knn = accuracy_score(y_test, y_pred_knn)
print(f"kNN Accuracy: {accuracy_knn:.4f}")
print("Classification Report:\n", classification_report(y_test, y_pred_knn))

# Step 5: Confusion Matrix
cm_knn = confusion_matrix(y_test, y_pred_knn)
print("Confusion Matrix:\n", cm_knn)

# Step 6: Plot Confusion Matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm_knn, display_labels=knn.classes_)
disp.plot(cmap=plt.cm.Blues)
plt.title("kNN - Confusion Matrix (SMOTE Applied)")
plt.grid(False)
plt.show()
```

```python
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import pandas as pd

# Step 1: Apply SMOTE on training data
smote = SMOTE(random_state=42)
X_train_sm, y_train_sm = smote.fit_resample(X_train, y_train)

# Convert both training and test data to NumPy arrays
X_train_sm_np = X_train_sm.values
X_test_np = X_test.values
y_train_sm_np = y_train_sm.values # Also convert y_train_sm


# Step 2: Train XGBoost classifier
# Removed use_label_encoder and eval_metric for broader compatibility
xgb_model = XGBClassifier(
    random_state=42
)
xgb_model.fit(X_train_sm_np, y_train_sm_np)

# Step 3: Predictions on test set
y_pred_xgb = xgb_model.predict(X_test_np)

# Step 4: Accuracy score
accuracy_xgb = accuracy_score(y_test, y_pred_xgb)
print(f"XGBoost Accuracy with SMOTE: {accuracy_xgb:.4f}")

# Step 5: Confusion matrix
cm_xgb = confusion_matrix(y_test, y_pred_xgb)
print('Confusion Matrix:\n',cm_xgb)

# Step 6: Feature importance plot
# Need to use original column names for feature importance
feature_importance = pd.Series(xgb_model.feature_importances_, index=X_train.columns)
feature_importance = feature_importance.sort_values(ascending=False)

plt.figure(figsize=(10, 6))
feature_importance.head(15).plot(kind='barh', color='darkorange')
plt.title("XGBoost - Top 15 Feature Importances (with SMOTE)")
plt.xlabel("Importance Score")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
